```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
df = pd.read_csv("cv_vs_temp.csv")
```

```python
temp = df["temp"]
fig, axs = plt.subplots(figsize=(3.33, 3.33))
for x in df.columns[1:]:
    plt.plot(temp, df[x] + int(x))
    peaks, _ = find_peaks(df[x], height=0)
    plt.plot(temp.values[peaks], df[x].values[peaks] + int(x), "x")
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(3.33, 3.33))
for x in df.columns[1:]:
    peaks, _ = find_peaks(df[x], height=0)
    nat = np.repeat(int(x), peaks.shape[0])
    plt.plot(temp.values[peaks], nat, "kx")
plt.show()
```
