```python
import matplotlib.pyplot as plt
```

```python
root = "/Users/robertwexler/pymatnest/test/"
c1 = []
c2 = []
c4 = []
with open(root + "1-core/output.energies") as f:
    next(f)
    for line in f:
        c1.append(float(line.split()[1]))
with open(root + "2-cores/output.energies") as f:
    next(f)
    for line in f:
        c2.append(float(line.split()[1]))
with open(root + "4-cores/output.energies") as f:
    next(f)
    for line in f:
        c4.append(float(line.split()[1]))
```

```python
plt.plot(c1, label="laptop, 1 core")
plt.plot(c2, "--", label="laptop, 2 cores")
plt.plot(c4, ":", label="laptop, 4 cores")
plt.ylim(-2, 10)
plt.legend()
plt.ylabel("Enthalpy (eV)")
plt.xlabel("NS iteration")
plt.savefig("enthalpy-vs-iteration.png")
```
